````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib inline
````

Create a normal distribution with mean as $\mu$ and standard deviation as $\sigma.$ We will use,
```python
mu = 1500
std = 300
snd = stats.norm(mu, std)
```

Generate 1000 random values between $a$, $b.$ We will know these limites after plotting.
```python
x = np.linspace(-500, 5000, 1000)
```

Plot the standard normal distribution for different values of random variable
falling in the range $a, b$

````python
def myNormalDistribution(mu, std):
    plt.figure(figsize=(8,4))
    x = np.linspace(100, 3000, 1000)
    snd = stats.norm(mu, std)
    plt.plot(x, snd.pdf(x))
    plt.xlim(200, 3000)
    plt.xlabel('Values of Random Variable X', fontsize='15')
    plt.ylabel('Probability', fontsize='15')
    plt.title(r'Normal Distribution', fontsize='15')
    plt.show();
````

````python
myNormalDistribution(1500, 300)
````

````python
interact(myNormalDistribution, mu=(800, 2000, 1), std=(100, 800, 0.1));
````

The ```cdf``` method calculates the proportion of a normally distributed population that is less than or equal to a given value. (CDF stands for Cumulative Distribution Function).

```cdf``` function calculates the probability for a given normal distribution value, and ```ppf()``` (percent point function) function calculates the normal distribution value for which a given probability is the required value. 

````python
def myCalculation(mu, std, x):
    snd = stats.norm(mu, std)
    print("CDF = ",snd.cdf(x), "PPF = ",snd.ppf(snd.cdf(x)))
````

````python
interact(myCalculation, mu=widgets.FloatSlider(value=1500, min=800, max=2000, step=1), 
         std=widgets.FloatSlider(value=300, min=100, max=800, step=1), 
         x =widgets.FloatSlider(value= 1800, min=500, max=2500,step=1));
````

````python
def myCDF(SupLimit):
    snd = stats.norm(1500, 300)
    plt.figure(figsize=(12, 6))
    x = np.linspace(100, 3000, 1000)
    shaded_region =  (x <= SupLimit)
    plt.fill_between(x, snd.pdf(x), alpha=0.2)
    plt.shaded_region =  (x <= SupLimit)
    plt.fill_between(x, snd.pdf(x), where=shaded_region)
    plt.xlim(200, 3000)
    plt.title(r'Normal Distribution $(\mu= 1500, \sigma = 300)$', fontsize='15')
    plt.xlabel('Values of Random Variable X', fontsize='15')
    plt.ylabel('CDF', fontsize='15');
    plt.show();
````

````python
interact(myCDF, SupLimit = 1800)
````
