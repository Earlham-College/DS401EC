# Diagonostics

To assess whether the linear model is reliable, we need to check for meeting three conditions: Linearity, Nearly normal residual and constant variability. 

1. **Linearity:** You already checked if the relationship between runs and at-bats is linear using a scatterplot. We should also verify this condition with a plot of the residuals vs. at-bats.
```python
sns.residplot(x, y, color = 'red')
plt.xlabel('at_bats', fontsize = 16)
plt.ylabel('residuals', fontsize = 16)
plt.show();
```
If the residuals exhibit a non-random pattern, such as a curve, it suggests that the relationship between the variables may be non-linear [0]. 


2. **Nearly normal residuals** We can plot a Histogram [1]

````python
residuals = y - y_pred
plt.hist(residuals, bins = 6)
plt.show();
````

Or we can plot a normal probability plot of the residuals. Use ```probplot``` from ```scipy.stats``` module.

````python
probplot(residuals, plot = plt)
plt.show(); 
````

3 **constant variability** It refers to the assumption of constant variance of the errors, also known as homoscedasticity. This assumption is one of the key assumptions of linear regression, and it means that the variance of the error terms (residuals) is the same for all values of the independent variable(s).

````python
plt.scatter(residuals, y_pred);
plt.xlabel('Residual', fontsize = 16)
plt.ylabel('Fitted values', fontsize = 16);
plt.show(); 
````
